```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
max_features = 10000
(xtrain,ytrain),(xtest,ytest)=imdb.load_data(num_words=max_features)
print(xtrain.shape)
print(xtest.shape)
```

```python
sample_review=xtrain[0]
sample_label=ytrain[0]
sample_review
sample_label
```

```python
word_index=imdb.get_word_index()
word_index
reverse_word={value:key for key,value in word_index.items()}
reverse_word
```

```python
max_len=500
xtrain=sequence.pad_sequences(xtrain,maxlen=max_len)
xtrain=sequence.pad_sequences(xtest,maxlen=max_len)
xtrain
```

```python
xtest
```

```python
xtrain[0]
```

```python
model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len))
model.add(SimpleRNN(128,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
model.fit(
    xtrain,ytrain,epochs=10,batch_size=32,
    validation_split=0.2,
    callbacks=[earlystopping]
)
```

```python
model.save('simple_rnn_imdb.h5')
```
